301: Auditoría de Integridad y Estructura de Origen

Este bloque realiza una disección técnica del dataset masivo, detectando inconsistencias de tipos y vacíos de información que podrían comprometer la solvencia del modelo predictivo.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estándar ITBA / Clase Mundial
plt.style.use('seaborn-v0_8-whitegrid')
pd.options.display.max_columns = None

# ===========================================================================
# 301.1 CARGA CRÍTICA Y SANITIZACIÓN
# ===========================================================================
# Cargamos el universo total para mapear la pérdida de información desde el origen
path_raw = r'../raw_data/nyswcb_claims.csv'
df = pd.read_csv(path_raw, low_memory=False)

# Sanitización de cabeceras: Eliminamos ruidos de caracteres no imprimibles (\r, \n)
# Un experto no permite nombres de columnas con basura de sistema
df.columns = [c.strip().replace('\r', '').replace('\n', '') for c in df.columns]

# ===========================================================================
# 301.2 AUDITORÍA DE CALIDAD DE DATOS (Data Hygiene)
# ===========================================================================
# Calculamos la completitud de las variables para justificar la selección de features
audit_report = pd.DataFrame({
    'Dtype': df.dtypes,
    'Nulls': df.isnull().sum(),
    'Null_%': (df.isnull().sum() / len(df)) * 100,
    'Unique_Values': df.nunique()
}).sort_values('Null_%', ascending=False)

print("--- REPORTE TÉCNICO DE INTEGRIDAD SOBERANA (N=3,129,701) ---")
print(audit_report.head(20))

# ===========================================================================
# 301.3 VISUALIZACIÓN DE LA DENSIDAD DE INFORMACIÓN
# ===========================================================================

plt.figure(figsize=(12, 6))
audit_report['Null_%'].head(25).plot(kind='bar', color='firebrick', alpha=0.8)
plt.axhline(y=30, color='black', linestyle='--', label='Umbral Crítico de Imputación (30%)')
plt.title('Auditoría de Vacíos de Información por Variable (Perspectiva de Higiene)', fontsize=14)
plt.ylabel('Porcentaje de Nulos (%)')
plt.xlabel('Variables del Diccionario WCB')
plt.legend()
plt.tight_layout()
plt.show()
```
